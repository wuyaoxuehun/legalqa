```python
import json
data = []
files = [f'./data/1_train.json']
# files = [f'./data/1_train.json']
for file in files:
    with open(file, 'r', encoding='utf8') as f:
        for line in f:
            example = json.loads(line)
            data.append(example)
```

```python
def display(examples):
    for example in examples:
        for k,v in example.items():
            if type(v) == type({}):
                display([v])
            else:
                print(k,v)

def filter_states(statement):
    import re
    expression = r'(哪一|哪种|哪个)'
    if re.search(expression, statement):
        return True
    else:
        return False

filter_states("以下选项哪项是正确的？")
```

```python
data_single = list(filter(lambda x:len(x['answer']) == 1, data))
data_multi = list(filter(lambda x:len(x['answer']) != 1, data))
print(len(data_single))
print(len(data_multi))
fun = lambda x : not filter_states(x['statement'])
data_single = list(filter(fun, data_single))
data_multi = list(filter(lambda x:not fun(x), data_multi))
data_combine = list(filter(lambda x:x['statement'].find('(1)')!=-1, data))
print(len(data_single))
print(len(data_multi))
# print(display(data))
# print(display(data_combine))
print(display(data_single))
# print(display(data_multi))
```
